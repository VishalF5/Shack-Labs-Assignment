Importing Libraries

```python
import pandas as pd
import tensorflow_hub as hub
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Amazon data
amazon = pd.read_csv('/content/drive/MyDrive/amz_com-ecommerce_sample.csv' ,encoding = 'iso-8859-1')
```

```python
# Flipkart data
flipkart = pd.read_csv('/content/drive/MyDrive/flipkart_com-ecommerce_sample.csv' ,encoding = 'iso-8859-1')
```

Filtering Columns

```python
amazon = amazon[['product_name','retail_price','discounted_price']]
flipkart = flipkart[['product_name','retail_price','discounted_price']]
```

Encoding Text

```python
encoder_url = 'https://tfhub.dev/google/universal-sentence-encoder/4' 

encoder = hub.load(encoder_url)
```

```python
encode_amazon = encoder(amazon.product_name)
encode_flipkart = encoder(flipkart.product_name)
```

Calculating Encoded Text's Similarity using Cosine Similarity

```python
similarity = cosine_similarity(encode_amazon, encode_flipkart)
```

```python
similarity
```

```python
# Function for finding similar products between two datasets.

def product_match(product):

    # Index on product name
    index = flipkart[flipkart['product_name'] == product].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])

    # Flipkart data columns
    flip_name =[]
    flip_retail_price = []
    flip_discount_price = []


    # Amazon data columns
    amz_name = []
    amz_retail_price = []
    amz_discount_price = []

    # Appending first 5 similar products information
    for i in distances[1:6]:
        flip_name.append(flipkart.iloc[i[0]].product_name)
        flip_retail_price.append(flipkart.iloc[i[0]].retail_price)
        flip_discount_price.append(flipkart.iloc[i[0]].discounted_price)


        amz_name.append(amazon.iloc[i[0]].product_name)
        amz_retail_price.append(amazon.iloc[i[0]].retail_price)
        amz_discount_price.append(amazon.iloc[i[0]].discounted_price)

    # Creating Dataframe 
    final = pd.DataFrame([flip_name ,flip_retail_price ,flip_discount_price, amz_name ,amz_retail_price ,amz_discount_price])
    final = final.transpose() #To Transpose and make each rows as columns

    final.columns = ['Product name in Flipkart' ,'Retail Price in Flipkart' ,'Discounted Price in Flipkart' ,
                     'Product name in Amazon','Retail Price in Amazon' ,'Discounted Price in Amazon']
    return final
```

```python
product_match("Alisha Solid Women's Cycling Shorts")
```
